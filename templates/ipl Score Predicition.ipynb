# Spliting Data Set

```python
X_train = encoded_train.drop(labels='total', axis=1)[encoded_train['date'].dt.year <= 2016]
X_test = encoded_train.drop(labels='total', axis=1)[encoded_train['date'].dt.year >= 2017]
Y_train = encoded_train[encoded_train['date'].dt.year <= 2016]['total'].values
Y_test = encoded_train[encoded_train['date'].dt.year >= 2017]['total'].values
X_train.drop(labels='date', axis=True, inplace=True)
X_test.drop(labels='date', axis=True, inplace=True)

print("Training set: {} and Test set: {}".format(X_train.shape, X_test.shape))
```

## Random Forest


```python
from sklearn.ensemble import RandomForestRegressor
random_reg = RandomForestRegressor(verbose=1 )
random_reg.fit(X_train,Y_train)
```

```python
Y_pred = random_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, accuracy_score
# Random Forest Regression - Model Evaluation
print("---- Random Forest Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(Y_test, Y_pred)))
print("Mean Squared Error (MSE): {}".format(mse(Y_test, Y_pred)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(Y_test, Y_pred))))
```

## DECISION TREE

```python
# Decision Tree Regression Model
from sklearn.tree import DecisionTreeRegressor
decision_regressor = DecisionTreeRegressor()
decision_regressor.fit(X_train,Y_train)
```

```python
y_pred = decision_regressor.predict(X_test)
```

```python
print("---- Decison Tree Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(Y_test, y_pred)))
print("Mean Squared Error (MSE): {}".format(mse(Y_test, y_pred)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(Y_test, y_pred))))
```

## LINEAR REGRESSION

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,Y_train)
```

```python
y_pred2 = linear_regressor.predict(X_test)
```

```python
print("---- Linear Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(Y_test, y_pred2)))
print("Mean Squared Error (MSE): {}".format(mse(Y_test, y_pred2)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(Y_test, y_pred2))))
```
